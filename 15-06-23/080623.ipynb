```python
# Chrome Driver' ykleme ve ayarlar
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
driver.set_window_size(1250, 740)
driver.set_window_position(0, 0)

# Ana link ve link listesi dosyas
mainLink = "https://www.arabam.com/ikinci-el"
linklist_file = "linklist.txt"

# Link listesini dosyadan okuma veya bo liste olarak balatma
if os.path.exists(linklist_file):
    with open(linklist_file, "r") as file:
        linklist = file.read().splitlines() #linklist.txt dosyasn okuyarak her bir satr ayr bir e olarak ieren bir liste oluturur.
else:
    linklist = []

# Linkleri ekme ve link listesine ekleme
def get_links(driver, xpath):
    a = driver.find_element(By.XPATH, xpath)
    lnks = a.find_elements(By.TAG_NAME, "a")
    for lnk in lnks:
        href = lnk.get_attribute("href")
        if href and "https://www.arabam.com/oto-ekspertiz" not in href and "https://www.arabam.com/turbolar" not in href:
            if href not in linklist:
                linklist.append(href)

    with open(linklist_file, "w") as file:
        file.write("\n".join(linklist))

    return linklist

# lk sayfay ziyaret et
url = mainLink
driver.get(url)
sleep(4)

# Sonraki sayfalara tklayarak linkleri ekme
while True:
    next_buttons = driver.find_elements(By.XPATH, '//*[@id="pagingNext"]')

    if len(next_buttons) > 0:
        next_button = next_buttons[0]

        if next_button.is_enabled():
            next_button.click()
            sleep(10)
            linklist = get_links(driver, '//*[@id="js-hook-missing-space-content"]/div[2]/div/div[2]')
        else:
            break
    else:
        break

print(linklist)
print(len(linklist))
```

```python

processed_data_file = "processed_data.txt"
processed_data = []

for idx, item in enumerate(cardetail):
    data = item[1].split('\n')
    vehicle = {}
    
    # lgili linki ekleyin
    vehicle['Link'] = linklist[idx]

    # Fiyat ve konumu ayrma
    fiyat = data[0].split('\n')[0]
    fiyat = fiyat.replace('.', '').replace(' TL', '')  # '.' ve 'TL' kaldrlyor
    vehicle['Fiyat'] = int(fiyat)  # Fiyat integer olarak kaydediliyor

    konum = None

    # Konumu bulma
    for i in range(1, len(data)-1):
        if data[i] != 'Tat Kredini Hesapla':
            konum = data[i]
            break

    vehicle['Konum'] = konum

    # Dier zellikleri ayrma
    for i in range(1, len(data)-1, 2):
        if data[i] == 'Tat Kredini Hesapla':
            continue

        key = data[i].strip(':')
        value = data[i+1]
        
        # lgili karakter dnmleri yaplyor
        key = key.replace('', 'i').replace('', 'i').replace('', 'o').replace('', 'u').replace('', 's').replace('', 'c')
        key = key.replace('', 'C').replace('', 'S').replace('', 'U').replace('', 'O')
        key = key.replace(' ', '_').replace('','g').replace('','G').replace('-','_')

        if '.' in value:  # '.' kaldrlyor
            value = value.replace('.', '')
        
        if '-' in value:  # '.' kaldrlyor
            value = value.replace('-', '')
            
        if ',' in value:  # '.' kaldrlyor
            value = value.replace(',', '.')
            
        #if 'hp' in value:  # 'hp' kaldrlyor
        #   value = value.replace(' hp', '')
            
        #if 'cc' in value:  # 'cc' kaldrlyor
        #    value = value.replace('cc', '')

        #if 'km' in value:  # 'km' kaldrlyor
        #    value = value.replace('km', '')
        #if 'cm3' in value:
        #    value = value.replace('cm3','')
            
        if key == 'Ort._Yakit_Tuketimi' or key == 'Yakit_Deposu':
            if value is not None:
                value = value.replace('lt', '')
            else:
                value = None
        if key == 'Motor_Hacmi' or key == 'Motor_Gucu':
            if value is not None:
                value = value.replace('lt', '').replace('hp','').replace('cc','').replace('cm3','')
            else:
                value = None
        if key == 'Kilometre':
            if value is not None:
                value = value.replace('km', '').replace('kilometre','').replace('KM','')
            else:
                value = None

        if key in ['Yil', 'Kilometre']:  # Yl ve Kilometre integer olarak kaydediliyor
            value = int(value)
        if key in ['Motor_Hacmi', 'Ort._Yakit_Tuketimi', 'Motor_Gucu','Yakit_Deposu']:
            if value.strip():  # Deerin bo olmadndan emin olunuyor
                try:
                    value = float(value)
                except ValueError:
                    value = None  # Saysal bir deere dntrlemiyorsa 'None' olarak ayarlanyor
            else:
                value = None
            
        vehicle[key] = value

    processed_data.append(vehicle)

    

# lenmi verileri dosyaya yazma
with open(processed_data_file, "w", encoding="utf-8") as file:
    for vehicle in processed_data:
        for key, value in vehicle.items():
            file.write(f"{key}: {value}\n")
        file.write("\n")

print("Veriler processed_data.txt dosyasna kaydedildi.")
# lenmi verileri grntle
for vehicle in processed_data:
    print(vehicle)
```

```python
from sklearn.impute import KNNImputer

for column in _df.columns:
    if _df[column].dtype == 'object':
        most_frequent_value = _df[column].mode().values[0]
        _df[column].fillna(most_frequent_value, inplace=True)
    else:
        imputer = KNNImputer(n_neighbors=5)
        column_data = _df[column].values.reshape(-1, 1)
        _df[column] = imputer.fit_transform(column_data)
#EGER BOS DEGERIN SUTUN TIPI OBJECT ISE EN SIK TEKRAR EDEN VERI LE DOLDUR EGER SAYISAL SE KNN ALGORTMASI LE DOLDUR.
```
